```python
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
sns.FacetGrid(train, hue="Click", size=3) \
   .map(plt.scatter, "Position") \
   .add_legend()
plt.show()
```

```python
sns.FacetGrid(train, hue="Impression", size=6) \
   .map(sns.kdeplot, "Depth") \
   .add_legend()
plt.show()
```

```python
import leather


dot_data = train.columns

def colorizer(d):
    return 'rgb(%i, %i, %i)' % (d.x, d.y, 150)

chart = leather.Chart('Colorized dots')
chart.add_dots(dot_data, fill_color=colorizer)
chart.to_svg('examples/charts/colorized_dots.svg')
```
